```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
t = np.linspace(0, 13.8, 1000)  # Time from 0 to 13.8 billion years
scale_factor = np.exp(t)         # Scale factor of the universe (exponential expansion)
temperature = 1 / t              # Temperature of the universe (decreasing over time)

# Plot the scale factor
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(t, scale_factor)
plt.xlabel('Time (billion years)')
plt.ylabel('Scale Factor')
plt.title('Expansion of the Universe')

# Plot the temperature
plt.subplot(1, 2, 2)
plt.plot(t, temperature)
plt.xlabel('Time (billion years)')
plt.ylabel('Temperature')
plt.title('Cooling of the Universe')

plt.tight_layout()
plt.show()
```

```python
def big_bang_theory():
    # Constants
    universe_age_years = 13.8e9  # Age of the universe in years
    light_speed_kmps = 299792.458  # Speed of light in kilometers per second

    # Calculate the distance light has traveled since the Big Bang
    light_distance_km = universe_age_years * light_speed_kmps * 365.25 * 24 * 3600

    # Print a summary
    print("The Big Bang Theory:")
    print("- The universe began as a singularity around 13.8 billion years ago.")
    print("- It was an endlessly hot and dense point of energy.")
    print("- Galaxies, stars, and planets formed as the universe cooled and expanded.")
    print("- Cosmic microwave background radiation is the leftover heat from the initial explosion.")
    print("- Abundance of light elements (like hydrogen and helium) is also explained by the Big Bang.")
    print(f"\nDistance light has traveled since the Big Bang: {light_distance_km:.2f} kilometers")

# Run the program
big_bang_theory()
```
